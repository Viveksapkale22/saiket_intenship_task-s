### Task 3: Customer Segmentation

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Sem-8 work/Intenship_Work/saiket_intenship_selected_1monthvirtualinten/cleaned_churn_data.csv')

df.head()
```

STEP 1. Segmentation Techniques

```python
df['Tenure_Segment'] = pd.cut(
    df['tenure'],
    bins=[0, 12, 36, 72],
    labels=['New (0–1 yr)', 'Medium (1–3 yrs)', 'Long (3+ yrs)']
)
```

```python
df['Charge_Segment'] = pd.qcut(
    df['MonthlyCharges'],
    q=3,
    labels=['Low', 'Medium', 'High']
)
```

```python
df['Contract_Type'] = 'Month-to-month'
df.loc[df['Contract_One year'] == 1, 'Contract_Type'] = 'One year'
df.loc[df['Contract_Two year'] == 1, 'Contract_Type'] = 'Two year'
```

```python
segment_table = df.groupby(
    ['Tenure_Segment', 'Charge_Segment', 'Contract_Type']
).size().reset_index(name='Customer_Count')

segment_table.head(10)
```

Step 2. Understanding Customer Behavior

3.1 Churn Rate by Tenure Segment

```python
tenure_behavior = df.groupby('Tenure_Segment').agg(
    Total_Customers=('Churn', 'count'),
    Churned_Customers=('Churn', 'sum'),
    Churn_Rate=('Churn', 'mean')
).reset_index()

tenure_behavior['Churn_Rate'] *= 100
tenure_behavior
```

3.2 Churn Rate by Charge Segment

```python
charge_behavior = df.groupby('Charge_Segment').agg(
    Total_Customers=('Churn', 'count'),
    Churned_Customers=('Churn', 'sum'),
    Churn_Rate=('Churn', 'mean')
).reset_index()

charge_behavior['Churn_Rate'] *= 100
charge_behavior
```

3.3 Churn Rate by Contract Type

```python
contract_behavior = df.groupby('Contract_Type').agg(
    Total_Customers=('Churn', 'count'),
    Churned_Customers=('Churn', 'sum'),
    Churn_Rate=('Churn', 'mean')
).reset_index()

contract_behavior['Churn_Rate'] *= 100
contract_behavior
```

Step 3. Churn Analysis Within Combined Segments

```python
segment_churn = df.groupby(
    ['Tenure_Segment', 'Charge_Segment', 'Contract_Type']
).agg(
    Customers=('Churn', 'count'),
    Churn_Rate=('Churn', 'mean')
).reset_index()

segment_churn['Churn_Rate'] *= 100
segment_churn.sort_values('Churn_Rate', ascending=False).head(10)
```

Step 4. Identifying High-Value Customers

```python
high_value_customers = df[
    (df['Charge_Segment'] == 'High') &
    (df['Tenure_Segment'] == 'Long (3+ yrs)')
]


high_value_summary = high_value_customers.groupby('Contract_Type').agg(
    Customers=('Churn', 'count'),
    Churn_Rate=('Churn', 'mean')
).reset_index()

high_value_summary['Churn_Rate'] *= 100
high_value_summary
```
