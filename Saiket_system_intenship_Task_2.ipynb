```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Sem-8 work/Intenship_Work/saiket_intenship_selected_1monthvirtualinten/cleaned_churn_data.csv')

df.head()
```

```python
# Overall churn rate
churn_rate = df['Churn'].mean()

print(f"Overall Churn Rate: {churn_rate:.2%}")
```

```python
import matplotlib.pyplot as plt

# 1. Prepare the data (assuming 'df' is your loaded DataFrame)
churn_counts = df['Churn'].value_counts()

# 2. Create the plot
plt.bar(churn_counts.index.map({0: 'No Churn (Stayed)', 1: 'Churn (Left)'}), churn_counts.values, color=['#2ecc71', '#e74c3c'])

# 3. Add labels and title
plt.title('Overall Customer Churn Distribution', fontsize=14)
plt.xlabel('Customer Status', fontsize=12)
plt.ylabel('Count', fontsize=12)

# 4. Optional: Add a grid for easier reading
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 5. Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. Prepare the data
gender_count = df['gender'].value_counts()
print(gender_count)
# 2. Create the plot
plt.bar(gender_count.index.map({0: 'female', 1: 'male'}), gender_count.values, color=['lightblue', 'pink'])

# 3. Add labels and title
plt.title('Overall Customer gender Distribution', fontsize=14)
plt.xlabel('Customer gender Status', fontsize=12)
plt.ylabel('Count', fontsize=12)

# 4. Optional: Add a grid for easier reading
plt.grid(axis='y', linestyle='--', alpha=0.6)

# 5. Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create the matrix (crosstab)
# This counts how many people fall into each Gender + Churn bucket
matrix = pd.crosstab(df['gender'], df['Churn'])

# 2. Rename for clarity
matrix.index = ['Female', 'Male']
matrix.columns = ['Stayed', 'Left']

# 3. Plot as a Heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(matrix, annot=True, fmt='d', cmap='YlGnBu')

plt.title('Gender vs. Churn "Matrix"', fontsize=14)
plt.xlabel('Customer Action', fontsize=12)
plt.ylabel('Gender', fontsize=12)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for a professional look
sns.set_theme(style="whitegrid")


# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Churn', y='tenure', data=df, palette='husl')

# Adding descriptive labels
plt.title('Comparison of Tenure Distribution: Churn vs. No Churn', fontsize=14)
plt.xlabel('Customer Status', fontsize=12)
plt.ylabel('Standardized Tenure (Mean = 0)', fontsize=12)

# Change 0/1 to readable labels on the x-axis
plt.xticks([0, 1], ['No Churn (Stayed)', 'Churn (Left)'])

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create a single 'Contract_Type' column for the plot
# We assume 'Month-to-month' is the case where both others are 0
df['Contract_Type'] = 'Month-to-month'
df.loc[df['Contract_One year'] == 1, 'Contract_Type'] = 'One year'
df.loc[df['Contract_Two year'] == 1, 'Contract_Type'] = 'Two year'

# 2. Use catplot to create the faceted grid
# col='Contract_Type' creates a new chart for each type
g = sns.catplot(
    data=df,
    x='Churn',
    col='Contract_Type',
    kind='count',
    palette='viridis',
    height=5,
    aspect=0.8
)

# 3. Clean up the labels
g.set_axis_labels("Churn Status", "Number of Customers")
g.set_xticklabels(["Stayed (0)", "Left (1)"])
g.set_titles("{col_name} Contract")

# 4. Add value labels on each bar
for ax in g.axes.flatten():
    for container in ax.containers:
        ax.bar_label(container, label_type='edge', fontsize=10)


plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Re-group Payment Methods into one column
# We create a default 'Bank transfer' for the case where others are 0
df['Payment_Method'] = 'Bank transfer (automatic)'
df.loc[df['PaymentMethod_Credit card (automatic)'] == 1, 'Payment_Method'] = 'Credit card'
df.loc[df['PaymentMethod_Electronic check'] == 1, 'Payment_Method'] = 'Electronic check'
df.loc[df['PaymentMethod_Mailed check'] == 1, 'Payment_Method'] = 'Mailed check'

# 2. Create the faceted grid
# col='Contract_Type' splits the windows
# x='Payment_Method' puts the methods on the bottom
# hue='Churn' colors them by who stayed vs. left
g = sns.catplot(
    data=df,
    x='Payment_Method',
    hue='Churn',
    kind='count',
    palette='magma',
    height=5,
    aspect=1.2
)

# 3. Rotate labels so the payment names are readable
g.set_xticklabels(rotation=45)
g.set_axis_labels("Payment Method", "Number of Customers")
g.set_titles("{col_name} Contract")

plt.show()
```
