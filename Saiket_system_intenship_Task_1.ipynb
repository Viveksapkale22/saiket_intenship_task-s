3) Handling missing values: use pandas and numpy to find missing values

- I got to know issue when i see the TotalCharges      7043 non-null   object
- Check the Tenure so it is zero
-  Handling missing by removing it.


```python
print("Tenure zero:", (df["tenure"] == 0).sum())
print("TotalCharges empty:", (df["TotalCharges"] == ' ').sum())
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(' ', np.nan)
df["tenure"] = pd.to_numeric(df["tenure"], errors="coerce")
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"], errors="coerce")
df.loc[df["tenure"] == 0, "tenure"] = np.nan

print("TotalCharges nan:", df["TotalCharges"].isna().sum())
print("tenure nan:", df["tenure"].isna().sum())
```

```python

df.dropna(subset=["tenure", "TotalCharges"], inplace=True)
print("TotalCharges nan:", df["TotalCharges"].isna().sum())
print("tenure nan:", df["tenure"].isna().sum())
```

```python
# Handling missing values

ndf = df.isnull().sum()
print(ndf)
```

```python
df = df.drop('customerID', axis=1)
```

5) Categorical variable encoding
- binary encoding
- onehot encoding

```python

binary_cols = [
    'gender',
    'Partner',
    'Dependents',
    'PhoneService',
    'PaperlessBilling',
    'Churn'
]
binary_map = {
    'Yes': 1,
    'No': 0,
    'Male': 1,
    'Female': 0
}

for col in binary_cols:
    df[col] = df[col].map(binary_map)

df.head()
```

```python

onehot_cols = [
    'OnlineBackup',
    'MultipleLines',
    'InternetService',
    'OnlineSecurity',
    'DeviceProtection',
    'TechSupport',
    'StreamingTV',
    'StreamingMovies',
    'Contract',
    'PaymentMethod'
]

# Filter onehot_cols to only include columns that are currently of object or category dtype
# This helps prevent errors if some columns were already converted to numeric by other cells
columns_to_onehot = [col for col in onehot_cols if col in df.columns and df[col].dtype in ['object', 'category']]

df = pd.get_dummies(df, columns=columns_to_onehot, drop_first=True)

df.head()
```
